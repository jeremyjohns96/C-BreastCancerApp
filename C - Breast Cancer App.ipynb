# Breast cancer App #

Welcome to the Breast Cancer App. To use this app follow these instructions. f measure the radius of all the concave points. Take the largest radius is the 'concave points_worst'. Take the mean of all the concave points to get 'concave points_mean'. 
Measure the worst radius by measuring from the center of the tumour to the furthest edge of the tumor. Take that as the 'radius worst'.
Enter these values in the code below replacing the label marked '5'. 
The algorithm will predict if the tumour is cancerous or benign with atleast a 92% accuracy. A 1 represents a Malignant tumor. A 0 reqpresents a Benign tumor.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
data = pd.read_csv("data.csv")
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics 
data['diagnosis2'] = np.where(data['diagnosis']=="M",1,0)
data2 = data[['concave points_mean','radius_worst','concave points_worst']]
target2 = data[['diagnosis2']] 
clf = DecisionTreeClassifier()
clf = clf.fit(data2[0:400],target2[0:400])
pred2 = clf.predict(data2[401:])
print("Accuracy:",metrics.accuracy_score(target2[401:],pred2))
corr = data.corr()
mask = np.triu(np.ones_like(corr,dtype=bool))
f,ax = plt.subplots(figsize=(11,9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Enter your values here
# Eg: Use 1.35, 1.77, 2.8
print(clf.predict(pd.DataFrame({'concave points_mean':[5],'radius_worst':[5],'concave points_worst':[5]})))
```

## Additional Info ##

### Swarmplot of 'concave points_mean' ###

#### The difference between the range of concave points mean of maligned and benign tumors can be appreciated here ####

```python
mean0 = 0
mean1 = 0
count0 = 0
count1 = 0
for i in range(data.shape[0]):
    if (data['diagnosis2'][i] == 0):
        mean0 += data['concave points_mean'][i]
        count0 += 1
    elif (data['diagnosis2'][i] == 1):
        mean1 += data['concave points_mean'][i]
        count1 += 1

mean0 = mean0/count0
mean1 = mean1/count1
   

df = pd.DataFrame(data)
print("::Concave points mean::")
print("Benign:",mean0)
print("Malignant:",mean1)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.swarmplot(y="concave points_mean",x="diagnosis",data=data)
```

#### The difference between the range of the mean of radius of maligned and benign tumors can be appreciated here

```python
mean0 = 0
mean1 = 0
count0 = 0
count1 = 0
for i in range(data.shape[0]):
    if (data['diagnosis2'][i] == 0):
        mean0 += data['radius_worst'][i]
        count0 += 1
    elif (data['diagnosis2'][i] == 1):
        mean1 += data['radius_worst'][i]
        count1 += 1

mean0 = mean0/count0
mean1 = mean1/count1

print("::Mean of Radius worst::")
print("Benign:",mean0)
print("Malignant:",mean1)
sns.swarmplot(y="radius_worst",x="diagnosis",data=data)
```

```python
mean0 = 0
mean1 = 0
count0 = 0
count1 = 0
for i in range(data.shape[0]):
    if (data['diagnosis2'][i] == 0):
        mean0 += data['concave points_worst'][i]
        count0 += 1
    elif (data['diagnosis2'][i] == 1):
        mean1 += data['concave points_worst'][i]
        count1 += 1

mean0 = mean0/count0
mean1 = mean1/count1

print("::Mean of Concave points worst::")
print("Benign:",mean0)
print("Malignant:",mean1)
sns.swarmplot(y="concave points_worst",x="diagnosis",data=data)
```
